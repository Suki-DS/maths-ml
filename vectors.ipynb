```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

#### Plotting a vector

```python
# quiver plot is basically a 2D plot which shows a vector lines as arrow 
plt.quiver(0,0,4,5)
plt.show()
```

```python
# add proper scales and units
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
```

```python
# add proper scales and units
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
# changing the colour of the vector - blue
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
# plotting multiple vectors in single plot
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1, color='b')
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
# addition of two vectors 
# vectors are arrays, so we will use numpy for this
# np.asarray - Convert the input to an array.
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vector_2
print(sum)
```

```python
# plotting the above values 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1, color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1, color='b')
# plotting the result of sum 
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# subtraction of two vectors 
# vectors are arrays, so we will use numpy for this
# np.asarray - Convert the input to an array.
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
difference = vector_1 - vector_2
print(difference)
```

```python
# plotting the above values 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1, color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1, color='b')
# plotting the result of difference 
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# multiply a vector by a scalar (result would be a vector)
vector_1 = np.asarray([0,0,2,3])
vector_2 = 2*vector_1
print(vector_2)
```

```python
# plotting the above values 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1, color='b')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# multiply a vector by a scalar (result would be a vector)
vector_1 = np.asarray([0,0,2,3])
vector_2 = -0.5*vector_1
print(vector_2)
```

```python
# plotting the above values 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1, color='b')
plt.quiver(0,0,-1.0,-1.5,scale_units='xy',angles='xy',scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
